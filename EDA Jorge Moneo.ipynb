Hipotesis 2: Evolución de los tres sectores por número de trabajadores a lo largo de treinta años

```python
# Creamos variable para las columnas solo con los años

columnas_años = trabajadores_totales.columns[:-4]
columnas_años_agriculture = t_agriculture.columns[:-4]
columnas_años_industry = t_industry.columns[:-4]
columnas_años_service = t_service.columns[:-4]
```

```python
# Sumamos los trabajadores

suma_por_año = trabajadores_totales[columnas_años].sum(axis=0)
suma_por_año_agriculture = t_agriculture[columnas_años].sum(axis=0)
suma_por_año_industry = t_industry[columnas_años].sum(axis=0)
suma_por_año_service = t_service[columnas_años].sum(axis=0)
```

```python
# Hacemos que sean una lista los años

años = [int(columna) for columna in columnas_años]
años_agriculture = [int(columna) for columna in columnas_años_agriculture]
años_industry = [int(columna) for columna in columnas_años_industry]
años_service = [int(columna) for columna in columnas_años_service]
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=años, y=suma_por_año, color='r')
plt.xlabel('Año')
plt.ylabel('Trabajadores')
plt.title('Trabajadores por año')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=años_agriculture, y=suma_por_año_agriculture)
plt.xlabel('Año')
plt.ylabel('Trabajadores agrícolas')
plt.title('Trabajadores agrícolas por año')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=años_industry, y=suma_por_año_industry, color='orange')
plt.xlabel('Año')
plt.ylabel('Trabajadores industriales')
plt.title('Trabajadores industriales por año')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=años_service, y=suma_por_año_service, color='g')
plt.xlabel('Año')
plt.ylabel('Trabajadores servicios')
plt.title('Trabajadores servicios por año')
plt.xticks(rotation=45)
plt.show()
```

**Al contemplar los datos generales de todos los sectores unificados, se observa un aumento en la población trabajadora a nivel general. Este incremento no es sorprendente, dado que la población mundial está en aumento y esto es una consecuencia natural. En el sector agrícola, sin embargo, se observa un descenso en el número de trabajadores a pesar del crecimiento poblacional, lo que indica un cambio de dirección con respecto a la agricultura. Por otro lado, la industria experimenta un crecimiento constante. Por último, en el sector de servicios, se observa un crecimiento significativo, siendo el sector con mayor crecimiento. Esto sugiere que los sectores secundario y terciario están absorbiendo a los trabajadores que en el pasado habrían trabajado en el sector primario.**

```python
sectores_idh['Total'] = sectores_idh[['Muy Alto', 'Alto', 'Medio', 'Bajo']].sum(axis=1)

plt.figure(figsize=(10, 6))
plt.bar(sectores_idh['Sector'], sectores_idh['Muy Alto'], color='b', label='Muy Alto')
plt.bar(sectores_idh['Sector'], sectores_idh['Alto'], bottom=sectores_idh['Muy Alto'], color='orange', label='Alto')
plt.bar(sectores_idh['Sector'], sectores_idh['Medio'], bottom=sectores_idh['Muy Alto'] + sectores_idh['Alto'], color='g', label='Medio')
plt.bar(sectores_idh['Sector'], sectores_idh['Bajo'], bottom=sectores_idh['Muy Alto'] + sectores_idh['Alto'] + sectores_idh['Medio'], color='r', label='Bajo')

plt.xlabel('Sector')
plt.ylabel('Suma Último Año')
plt.title('Comparación de Sumas por Sector y Tipo de IDH')
plt.legend()
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

**Se contempla en qué medida el Índice de Desarrollo Humano (IDH) influye en la elección de la industria en la que te dedicas. Si se trata de un país con un IDH bajo, tienes más posibilidades de trabajar en una profesión agrícola. Si el IDH es medio, tienes las mismas posibilidades de trabajar tanto en la agricultura como en los servicios, aunque la industria adquiere un poco más de peso pero sigue siendo la menor. Al avanzar a un índice de desarrollo humano alto, la agricultura pierde peso en favor de la industria y los servicios crecen. Por último, al ser muy alto el IDH, la diferencia es mucho más dramática, siendo mínimo el sector agrícola frente a los servicios, y la industria estaría entre los dos. En resumen, se observa cómo a medida que el IDH es más alto, la agricultura pierde importancia en favor de la industria y, sobre todo, de los servicios.**
